### Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all?

A probability distribution is a mathematical function that describes the likelihood of all possible outcomes of a random variable in a given experiment or event. It is used to model the probabilities of various outcomes of a random process or event. Probability distributions can be continuous or discrete, and they can be represented using probability density functions or cumulative distribution functions.

Although the values in a probability distribution are random, we can still make predictions about their likelihood based on the probability distribution function. For example, if we have a fair six-sided die, the probability distribution for rolling any particular number is 1/6, or approximately 0.167. While we can't predict the exact outcome of rolling the die, we can predict the likelihood of each possible outcome based on the probability distribution.

Here's an example of creating and using a probability distribution in Python using the `scipy.stats` module:

```python
import scipy.stats as stats

# Create a normal distribution with mean=0 and standard deviation=1
norm_dist = stats.norm(0, 1)

# Calculate the probability density function for x=1
pdf_at_1 = norm_dist.pdf(1)

# Calculate the cumulative distribution function for x=2
cdf_at_2 = norm_dist.cdf(2)

# Generate 1000 random samples from the distribution
samples = norm_dist.rvs(size=1000)
``` 

In this example, we create a normal distribution with mean 0 and standard deviation 1, and then use the `pdf` method to calculate the probability density function at `x=1` and the `cdf` method to calculate the cumulative distribution function at `x=2`. Finally, we generate 1000 random samples from the distribution using the `rvs` method. While the specific values of the samples are random, their distribution will follow the probability distribution function we defined.


### Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?

Yes, there is a distinction between true random numbers and pseudo-random numbers. True random numbers are generated by natural sources of entropy, such as atmospheric noise or radioactive decay, and their values are completely unpredictable. On the other hand, pseudo-random numbers are generated by deterministic algorithms that use a seed value as an input, and their values are predictable given the same seed value.

Pseudo-random numbers are considered "good enough" for many applications because they exhibit statistical randomness properties that make them suitable for simulations, modeling, and other scientific purposes. The Python standard library provides several modules for generating pseudo-random numbers, including the `random` module.

Here's an example of generating a pseudo-random number in Python using the `random` module:

```python
import random

# Generate a random float between 0 and 1
rand_num = random.random()
print(rand_num)
```

Output:
```
0.3664207177054141
```


### Q3. What are the two main factors that influence the behaviour of a &quot;normal&quot; probability distribution?

The two main factors that influence the behavior of a "normal" probability distribution are the mean and the standard deviation. The mean determines the center of the distribution, while the standard deviation determines the spread of the distribution.

In Python, we can use the `numpy` library to generate a normal distribution with a specified mean and standard deviation. Here's an example:

```python
import numpy as np

# Generate 1000 random numbers from a normal distribution with mean 0 and standard deviation 1
data = np.random.normal(0, 1, 1000)

# Calculate the mean and standard deviation of the data
mean = np.mean(data)
std = np.std(data)

print("Mean:", mean)
print("Standard deviation:", std)
```

In this example, we generate 1000 random numbers from a normal distribution with a mean of 0 and a standard deviation of 1. We then calculate the mean and standard deviation of the resulting data. Because we specified a mean of 0 and a standard deviation of 1, we would expect the calculated mean and standard deviation to be close to 0 and 1, respectively.


### Q4. Provide a real-life example of a normal distribution. 

A real-life example of a normal distribution is the distribution of heights in a population. It is well-known that the heights of people in a population tend to follow a normal distribution, with most people being of average height and fewer people being either very tall or very short. 

Let's consider an example where we simulate the heights of people in a population using Python's NumPy library:


````python

import numpy as np
import matplotlib.pyplot as plt

# generate a normal distribution of heights with mean=170 and standard deviation=10
heights = np.random.normal(loc=170, scale=10, size=10000)

# plot the histogram of heights
plt.hist(heights, bins=50, density=True, alpha=0.7)
plt.xlabel('Height (cm)')
plt.ylabel('Probability density')
plt.show()
````

This code generates a normal distribution of heights with mean 170 cm and standard deviation 10 cm, and then plots the histogram of the generated heights. The resulting plot shows a bell-shaped curve, which is characteristic of a normal distribution. Most of the heights are clustered around the mean of 170 cm, and fewer heights are found as we move away from the mean.

### Q5. In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows?

In the short term, the behavior of a probability distribution can be unpredictable due to randomness. However, as the number of trials grows, the probability distribution is expected to converge to its expected or theoretical distribution.

For example, let's consider a coin toss experiment. If we toss a coin 10 times, we might observe an unequal number of heads and tails due to randomness. However, as we increase the number of tosses to 100, 1000, or more, we would expect the proportion of heads to tails to be closer to 0.5, which is the expected proportion for a fair coin. We can simulate this behavior using Python's NumPy library:


````python
import numpy as np
import matplotlib.pyplot as plt

# simulate coin tosses with different number of trials
trials = [10, 100, 1000, 10000]
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

for i, ax in enumerate(axs.flatten()):
    n = trials[i]
    tosses = np.random.binomial(1, 0.5, size=n)
    ax.hist(tosses, bins=2)
    ax.set_title(f'{n} Tosses')
    ax.set_xlabel('Outcome')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
````

This code simulates coin tosses with 10, 100, 1000, and 10000 trials and plots a histogram of the outcomes. As we can see from the resulting plots, the distribution of outcomes becomes closer to 0.5 as the number of trials increases.

### Q8. What is the relationship between exponentiation and logarithms?

The `math` package in Python provides functions for computing logarithms of different bases, including the natural logarithm (`log`) and the common logarithm (`log10`). Here are some examples:

```python

print(log_base_2_of_8)  # Output: 3.0

# Compute the natural logarithm of 1

# Compute the common logarithm of 100
common_log_of_100 = math.log10(100)
print(common_log_of_100)  # Output: 2.0
```

In the first example, we compute the logarithm base 2 of 8, which is equal to 3.0 since 2 raised to the power of 3 is 8. In the second example, we compute the natural logarithm of 1, which is equal to 0.0 since `e` raised to the power of 0 is 1. In the third example, we compute the common logarithm of 100, which is equal to 2.0 since 10 raised to the power of 2 is 100.


### Q9. What are the three logarithmic functions that Python supports?

Python's `math` module supports three logarithmic functions:

1. `math.log(x[, base])`: This function returns the natural logarithm of `x` (to the base `e`). If the optional argument `base` is provided, it returns the logarithm of `x` to the given `base`.

   Example:
   ```python
   >>> import math
   >>> math.log(10)

   ```

2. `math.log10(x)`: This function returns the base-10 logarithm of `x`.

   Example:
   ```python
   >>> import math
   >>> math.log10(100)
   2.0
   ```

3. `math.log2(x)`: This function returns the base-2 logarithm of `x`.

   Example:
